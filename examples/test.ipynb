```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# make 3 blobs
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
# to pandas dataframe
df = pd.DataFrame(X, columns=['x1', 'x2'])
```

```python
plt.scatter(df['x1'], df['x2'], s=50, cmap='viridis', c=y)
```

```python
shuffled = df.sample(frac=1)
```

```python
shuffled.index
```

```python
labels = KMeans(n_clusters=3, random_state=0).fit_predict(shuffled)
```

```python
plt.scatter(shuffled['x1'], shuffled['x2'], s=50, cmap='viridis', c=labels[shuffled.index])
```

```python
np.unique(labels[::-1], return_counts=True)
```

```python
np.bincount(labels[::-1])
```

```python
shuffled[shuffled["x1"] > shuffled["x2"]].index
```
